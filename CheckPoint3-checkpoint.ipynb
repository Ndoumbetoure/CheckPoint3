# 1.Preprocessing phase

```python
#data analysis and visualizagtion   

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
 
```

```python
titanic = pd.read_csv('titanic-passengers.csv',sep=";") #dta import

df = titanic.copy() #data copy
```

```python
print(titanic.columns.values)   #showing header
```

```python
titanic.head()    #show the general informations
```

```python
titanic.describe()  #data description
```

We can see that passenger ages range from 0.4 to 80.
And some features contain missing values, like the ‘Age’ feature.


```python
titanic.isnull().sum()  #find all missing values
```

```python
#replace missing Age values with median

titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)
```

```python
#replace missing Embarked values with mode
titanic['Embarked'].mode()
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
```

```python
#Dropping the Cabin columns

titanic = titanic.drop(columns='Cabin', axis=1)
```

```python
titanic.info()
```

The dataset has 891 rows, 11 columns  
2 of the features are floats, 5 are integers and 5 are objects

```python
#Transform the object type values to numerical values(survived).
titanic.replace({'Survived':{'No':0,'Yes':1}}, inplace=True)
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
   
```

```python
 plot_correlation_map(titanic)
```

# Interpretation

this graph is a correlation heatmap to understand the linear relationship between the different variables in our dataset.")

the graph shows us the strongest positive correlations in (pink) and the strongest negative correlations in (blue).

The first line contains the values that represent the correlation of each variable with the target variable (survived)

We can see that there are strong positive correlations between SibSp and Parch and this makes sense as parents and children often travel together.

We can also see that Pclass and Fare are strongly negatively correlated which makes sense as 1st class tickets are more expensive than lower class tickets,

Finally Age and Pclass are moderately negatively correlated - since wealthier people are generally older.



```python
#groupby between Pclass and Survived

titanic[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

# Interpretatuion

We observe a significant correlation (>0.5) between Pclass=1 and Survived.
Conclusion higher class passengers (Pclass = 1) were more likely to have survived.

```python
#drop the useless columns
titanic = titanic.drop(['PassengerId', 'Name', 'Ticket', 'Embarked'], axis=1)  
```

```python
#Extract Title from passager name

titanic['Title'] = df['Name'].map(\
    lambda name: name.split(',')[1]\
    .split('.')[0].strip()
)
```

```python
titanic.head(60)
```

```python
plotList = ['Title']

fig = plt.figure(figsize=(12,16))
plotNum  = 1     # initialize plot number
for i in plotList:
    fig.add_subplot(3,2,plotNum)
    ax = sns.countplot(x=sorted(titanic[i]),hue='Survived',data=titanic)
    plt.title('Survival Rate per \"{}\"'.format(i), fontsize=12)
    plt.xlabel(i, fontsize=12)
    plt.ylabel('Survival Rate', fontsize=12)
    plt.axis('auto')
    plt.legend(('died', 'survived'), loc='best')
  
plt.show()
```

```python
Title_Dictionary = {

                    "Capt":       "Officer",

                    "Col":        "Officer",

                    "Major":      "Officer",

                      "Dr":         "Officer",

                    "Rev":        "Officer",

                    "Jonkheer":   "Royalty",

                    "Don":        "Royalty",

                    "Sir" :       "Royalty",

                   "Lady" :      "Royalty",

                  "the Countess": "Royalty",

                    "Dona":       "Royalty",

                    "Mme":        "Miss",

                    "Mlle":       "Miss",

                    "Miss" :      "Miss",

                    "Ms":         "Mrs",

                    "Mr" :        "Mrs",

                    "Mrs" :       "Mrs",

                    "Master" :    "Master"

                    }
```

```python
titanic['Title'] = titanic.Title.map(Title_Dictionary)
titanic.head()
```

```python
#Transform the object type values to numerical values(Title).
encoder=LabelEncoder()
titanic['Title']=encoder.fit_transform(titanic['Title'])
titanic
```

```python
plotList = ['Title']

fig = plt.figure(figsize=(12,16))
plotNum  = 1     # initialize plot number
for i in plotList:
    fig.add_subplot(3,2,plotNum)
    ax = sns.countplot(x=sorted(titanic[i]),hue='Survived',data=titanic)
    plt.title('Survival Rate per \"{}\"'.format(i), fontsize=12)
    plt.xlabel(i, fontsize=12)
    plt.ylabel('Survival Rate', fontsize=12)
    plt.axis('auto')
    plt.legend(('died', 'survived'), loc='best')
  
plt.show()
```

```python
#create familysize
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] +1
titanic.head()
```

```python
#drop SbiSp and Parch columns
titanic = titanic.drop(['SibSp', 'Parch'], axis=1)
```

```python
titanic[['FamilySize', 'Pclass', 'Survived']].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Pclass', ascending=False)
```

```python
plotList = ['FamilySize']

fig = plt.figure(figsize=(12,16))
plotNum  = 1     # initialize plot number
for i in plotList:
    fig.add_subplot(3,2,plotNum)
    ax = sns.countplot(x=sorted(titanic[i]),hue='Survived',data=titanic)
    plt.title('Survival Rate per \"{}\"'.format(i), fontsize=12)
    plt.xlabel(i, fontsize=12)
    plt.ylabel('Survival Rate', fontsize=12)
    plt.axis('auto')
    plt.legend(('died', 'survived'), loc='best')
  
plt.show()
```

 Interpretation:
 
FamilySize shows us that families with four or fewer members had a better chance of surviving. Again, family size greater than four had a significant impact on survival chances. Passengers who were alone were more likely to die.
